```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,y=make_blobs(n_samples=100,n_features=2,centers=2,random_state=9)
y.reshape(y.shape[0],1)
```

```python
def initialisation(X):
    W=np.random.rand(X.shape[1],1)
    b=np.random.randn(1)
    return (W,b)
```

```python
def model(W,b,X):
    Z=np.dot(X,W)+b
    A=1/(1+np.exp(-Z))
    return A
```

```python
def log_loss(y,A):
    return 1/len(y)*(np.sum(-y*np.log(A)-(y-1)*np.log(1-A)))
```

```python
def gradients(X,A,y):
    dW=1/len(y)*np.dot(X.T,A-y)
    db=1/len(y)*np.sum(A-y)
    return (dW,db)
# def gradients(A,X,y):
#     dW=1/len(y)*np.dot(X.T,A-y)
#     db=1/len(y)*np.sum(A-y)
#     return (dW,db)
```

```python
def update(W,b,dW,db,lr):
    W=W-lr*dW
    b=b-lr*db
    return (W,b)
```

```python
def artificial_neurone(y,X,W,b,lr=0.01,n_iter=100):
    W,b=initialisation(X)
    loss=[]
    for i in range(n_iter):
        A=model(W,b,X)
        loss.append(log_loss(y,A))
        dW,db=gradients(X,A,y)
        W,b=update(W,b,dW,db,lr)
    plt.plot(loss)
    plt.show()
    return (W,b)
```

```python
W,b=artificial_neurone(y,X,W=0,b=0,lr=0.01,n_iter=100)
```
